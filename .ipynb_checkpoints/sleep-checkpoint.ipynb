### getting hold of sleep phases duration from individual days datasets ###

## Steps ##
1. import datasets 6-79
2. combine datasets
3. extract 5 columns: date, and 4 phases of sleep
4. combine datasets by adding 4 columns onto the sleep_data df based on common "Day" value

```python
# importing daily sleep datasets 
sleep_phases = pd.DataFrame()
for index in range(6,80):
    sindex = pd.read_csv(f"sleep ({index}).csv")
    
    if index == 6:
        sindex = sindex.T
        sleep_phases= pd.concat([sleep_phases, sindex], axis=0, ignore_index=True)
    else:
        data_column = pd.DataFrame(sindex.iloc[:, 1]).T
        sleep_phases= pd.concat([sleep_phases, data_column], axis=0, ignore_index=True)
```

```python
from datetime import datetime, time, timedelta
import re
```

```python
# function to convert date/time variables into a correct format for python 
def reformatting_time(time):
    time = str(time)
    hours = re.search(r'(\d+)h', time)
    minutes = re.search(r'(\d+)m', time)
    
    h = int(hours.group(1)) if hours else 0
    m = int(minutes.group(1)) if minutes else 0
    return timedelta(hours = h, minutes=m)
```

```python
sleep["DurationCF"]=sleep["Duration"].apply(reformatting_time)
```

```python
sleep["DeepSleepCF"]=sleep["DeepSleepDuration"].apply(reformatting_time)
sleep["LightSleepCF"]=sleep["LightSleepDuration"].apply(reformatting_time)
sleep["REMCF"]=sleep["REMDuration"].apply(reformatting_time)
sleep["AwakeTimeCF"]=sleep["Awake Time"].apply(reformatting_time)
```

```python
#need to convert sleep phases duration into a continuous variable
def to_continuous(time):
    time = str(time)
    hours = re.search(r'(\d+)h', time)
    minutes = re.search(r'(\d+)m', time)
    
    h = int(hours.group(1)) if hours else 0
    m = int(minutes.group(1)) if minutes else 0
    
    return h*60+m

sleep["DurationMin"] = sleep["Duration"].apply(to_continuous).replace(0,np.nan)
sleep["DeepSleepDurationMin"] = sleep["DeepSleepDuration"].apply(to_continuous).replace(0,np.nan)
sleep["LightSleepDurationMin"] = sleep["LightSleepDuration"].apply(to_continuous).replace(0,np.nan)
sleep["REMDurationMin"] = sleep["REMDuration"].apply(to_continuous).replace(0,np.nan)
sleep["AwakeTimeMin"] = sleep["Awake Time"].apply(to_continuous).replace(0,np.nan)
```
