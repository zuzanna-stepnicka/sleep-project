```python
# Importing packages
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# Import datasets

sleep_df1 = pd.read_json("data/2023-10-06_2024-01-14_118514764_sleepData.json")
sleep_df2 = pd.read_json("data/2024-01-14_2024-04-23_118514764_sleepData.json")
sleep_df3 = pd.read_json("data/2024-04-23_2024-08-01_118514764_sleepData.json")
sleep_df4 = pd.read_json("data/2024-08-01_2024-11-09_118514764_sleepData.json")
```

```python
sleep_df3.info()
```

```python
# Checking if all df are same, and merge

def df_check(df1, df2, df3,df4):

    # Chech if same number of columns
    if df1.shape[1]== df2.shape[1]== df3.shape[1]== df4.shape[1]:
        print("Same number of columns")
    else:
        print ("Coulmn numbers do not match")
    
    # Check if column names match
    if set(df1.columns) == set(df2.columns) == set(df3.columns) == set(df4.columns):
        print ("Same column names")
    else:
        print ("Coulmn names do not match")

     # Check if dtypes match
    if set(df1.dtypes) == set(df2.dtypes) == set(df3.dtypes) == set(df4.dtypes):
        print ("Same dtypes")
    else:
        print ("Dtypes do not match")
   

df_check(sleep_df1, sleep_df2, sleep_df3, sleep_df4)
```

```python
# Merge df
sleep_df = pd.concat([sleep_df1, sleep_df2, sleep_df3, sleep_df3], axis=0, ignore_index = True)
```

```python
sleep_df.info()
```

```python
sleep_df.T
```

```python
# Extract data from a dictionary in json file (i.e. "sleepScores") and simultaneusly join dataframes

sleep = pd.concat([sleep_df , pd.json_normalize(sleep_df["sleepScores"])], axis=1)
# check if columns number matches
if len(sleep.columns) == (len(sleep_df.columns)+len(sleep_df["sleepScores"][0].keys())):
    print("All good")
else:
    print("Something wrong")
```

```python
sleep = sleep.drop('sleepScores', axis=1)
```

```python
# Remove unwanted columns
print(sleep.columns)
```

```python
# retro?
print(sleep.retro.value_counts())
print(len(sleep))
```

```python
unwanted_columns = ['sleepWindowConfirmationType', 'feedback', 'insight', 'retro']
sleep=sleep.drop(columns = unwanted_columns, axis=1)
```

```python
sleep.dtypes
```

```python
# Change objects into datetime objects

start = pd.to_datetime(sleep['sleepStartTimestampGMT'][0], format = "%Y-%m-%dT%H:%M:%S.%f")
end = pd.to_datetime(sleep['sleepEndTimestampGMT'][0], format = "%Y-%m-%dT%H:%M:%S.%f")

duration = end - start
print(type(duration))
print(duration)
```

```python
sleep["BedTime"] = pd.to_datetime(sleep["sleepStartTimestampGMT"], format = "%Y-%m-%dT%H:%M:%S.%f")
sleep["WakeTime"] = pd.to_datetime(sleep["sleepEndTimestampGMT"], format = "%Y-%m-%dT%H:%M:%S.%f") 
sleep["SleepDuration"] = sleep["WakeTime"] - sleep["BedTime"]
sleep["CalendarDate"] = pd.to_datetime(sleep["calendarDate"], format = "%Y-%m-%d")
```

```python
sleep.dtypes
```

```python
#Convert SleepDuration variable into float
print(sleep.SleepDuration.dtype)
print(sleep.SleepDuration[0])
sleep["SleepDurationMin"] = sleep["SleepDuration"].dt.total_seconds()/60
print(sleep.SleepDurationMin[0].dtype)
```

```python
sleep= sleep.drop(['sleepStartTimestampGMT', 'sleepEndTimestampGMT', 'calendarDate'], axis=1)
```

```python
sleep.dtypes
```

```python
# Treat NA values
sleep.isna().sum()
```

```python
# Finding indexes of rows with NA values for columns that contain NAs
na_indexes = {col: set(sleep[sleep[col].isna()].index) for col in sleep.columns if sleep[col].isna().any()}
print(na_indexes)
```

```python
# Check if all sets of indexes are the same
if na_indexes['averageRespiration']== na_indexes['highestRespiration'] == na_indexes['lowestRespiration']:
    print ("Same indexes have missing values")
else:
    print("Different indexes have missing values")
```

```python
# Should I delete 26 rows from the dataset of imput the values as median?
print((len(sleep)-26)/len(sleep)*100)
```

```python
# For now I will remove rows with NA as its 9% of the data
sleep = sleep.dropna(axis = 0, how="any")
```

```python
# Saving cleaned dataframe for the future
sleep.to_csv("sleep_cleaned_df.csv", index=False)
```
